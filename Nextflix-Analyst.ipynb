```python
#Importing necessary libraries:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

#Supressing warnings:

import warnings
warnings.filterwarnings('ignore')
#reading data from csv file:
data = pd.read_csv('/content/drive/MyDrive/netflix_csv.csv')
data.info()
```

```python
#Checking the first 20 data from the dataframe
data.head(20)
```

```python
# dropping the rows having NaN values
df = data.dropna()
df.head(10)
```

```python
#Checking for null values

df.isnull().any()
df.isnull().sum()
```

```python
#Checking for unique values

df.nunique()
```

```python
# Question 1; What is the distribution of movies and TV shows in the dataset?

 # Count the number of movies and TV shows
type_counts = df['type'].value_counts()
piecolors=["#4CAF50","blue"]

 # Plot the distribution
plt.pie(type_counts, labels=type_counts.index, colors=piecolors, autopct='%1.1f%%')
plt.title('Distribution of Movies and TV Shows on Netflix')
plt.show()
```

```python
# Question 2; What are the most popular movies and TV shows in the dataset base on the ratings?

#group the dataset by title and calculate the mean rating for each title
title_ratings = df.groupby('title')['rating'].value_counts()

# sort the titles by mean rating in descending order
top_ratings = title_ratings.sort_values(ascending=False)

# display the top 10 titles with the highest mean ratings
print(top_ratings.head(10))
```

```python
#Question 3; How has the number of movies and TV shows added to Netflix changed over time?

 #group the dataset by type and release year and count the number of entries in each group
count_by_year = df.groupby([ 'type','release_year'])['show_id'].count()

# create a pivot table to reshape the data for plotting
count_by_year = count_by_year.reset_index().pivot(index='release_year', columns='type', values='show_id')

# plot the number of movies and TV shows added to Netflix over time
count_by_year.plot(kind='line', figsize=(10, 5))

# add a title and axis labels to the plot
plt.title('Number of Movies and TV Shows Added to Netflix Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')

# display the plot
plt.show()
```

```python
#Question4; What are the top countries of origin for movies and TV shows on Netflix?

#group the dataset by type and release year and count the number of entries in each group
count_by_country = df.groupby([ 'country','type'])['show_id'].count()

# create a pivot table to reshape the data for plotting
count_by_country = count_by_country.reset_index().pivot(index='country', columns='type', values='show_id')

# create a new column for the total number of entries for each country
count_by_country['total'] = count_by_country.sum(axis=1)

# sort the pivot table by the total number of entries in descending order
count_by_country = count_by_country.sort_values('total', ascending=False)

# select the top 10 countries by total number of entries
top_5_countries = count_by_country.head(5)

# display the top 10 countries
print(top_5_countries)
```

```python
#Question 5;What is the most common rating for movies and TV shows on Netflix?

#count the frequency of each rating
rating_counts = df['rating'].value_counts()

# select the rating with the highest frequency
most_common_rating = rating_counts.index[0]

print(rating_counts)
```

```python
 #Question 7; What are the top directors of movies and TV shows on Netflix?

# create a new DataFrame containing only movies and TV shows
movies_and_tv_shows = df[df['type'].isin(['Movie', 'TV Show'])]

# split the values in the 'director' column if there are multiple directors
directors = movies_and_tv_shows['director'].str.split(',', expand=True).stack().str.strip().reset_index(level=1, drop=True)

# count the occurrences of each director
top_directors = directors.value_counts().head(10)

print(top_directors)
```
